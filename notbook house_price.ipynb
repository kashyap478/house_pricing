```python
def remove_outliers(df):
    df_output =pd.DataFrame()
    for key,subdf in df.groupby('location'):
    # Calculate the mean and standard deviation
     mean =np.mean(subdf.price_per_sqrt)
     std = np.std(subdf.price_per_sqrt)

     gendf=subdf[((subdf.price_per_sqrt > (mean-std)) & (subdf.price_per_sqrt) <= (mean+std))]

     df_output= pd.concat([df_output,gendf],ignore_index=True)
    
    return df_output
```

```python
def bhk_outliers(df):
    exclude_indices =np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats ={}
        for bhk,bhk_df in location_df.groupby('location'):
            bhk_stats ={}   
            bhk_stats[bhk] ={
               'mean':np.mean(bhk_df.price_per_sqrt),
               'std': np.std(bhk_df.price_per_sqrt),
               'count': bhk_df.shape[0]}
            
        for bhk,bhk_df in location_df.groupby('location'):
            stats= bhk_stats.get(int(bhk-1))
            if stats and stats['count']>5: 
                exclude_indices =np.append(exclude_indices,bhk_df[bhk_df.price_per_sqrt <(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index') 

  
```

```python
def bhk_outliers(df):
    bhk_stats = {}  # To store stats for each BHK category
    exclude_indices = np.array([])

    # Calculate stats for each BHK group
    for bhk, bhk_df in df.groupby('bhk'):
        bhk_stats[bhk] = {
            'mean': np.mean(bhk_df.price_per_sqrt),
            'std': np.std(bhk_df.price_per_sqrt),
            'count': bhk_df.shape[0]
        }

    # For each BHK group, check if there are any outliers based on mean and std
    for bhk, bhk_df in df.groupby('bhk'):
        stats = bhk_stats.get(int(bhk)-1)  # Ensure `bhk` is treated as an integer
        if stats and stats['count'] > 5: 
            # Exclude rows where `price_per_sqrt` is below the threshold
            exclude_indices = np.append(
                exclude_indices, 
                bhk_df[bhk_df.price_per_sqrt < stats['mean']].index.values
            )

    return df.drop(exclude_indices)
```

```python
X=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Apply Linear Regression

```python
column_trans= make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr= LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

## applying Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

## Applying Ridge


```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

```python
print("No Regulation: ",r2_score(y_test,y_pred_lr))
print("Lasso: ",r2_score(y_test,y_pred_lasso))
print("Ridge: ",r2_score(y_test,y_pred_ridge))
```
